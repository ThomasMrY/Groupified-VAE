```python
import torch
x = torch.from_numpy(x/ 255.).permute(0,3,1,2).float()
```

```python
from torchvision.utils import make_grid
img = make_grid(x,nrow=8)
```

```python
from PIL import Image
def imshow(array):
    display(Image.fromarray(array))
```

```python
def rgb(t): return (np.clip((t[0] if len(t.shape) == 4 else t).detach().cpu().numpy().transpose([1, 2, 0]), 0, 1) * 255).astype(np.uint8)
```

```python
imshow(rgb(img))
```
